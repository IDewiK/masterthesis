### Assigning IRIs

```python
occupations_dict = {'': 'no Occupation'}
occ_idx = 0

#tempIndex = 0
#tempStrIndex = 0

df['IRI'] = ''
df['Occ_IRI'] = ''
df['Occ_1_IRI'] = ''
df['Occ_2_IRI'] = ''
df['Norm_Occ_IRI'] = ''
df['Norm_Occ_1_IRI'] = ''
df['Norm_Occ_2_IRI'] = ''

#Add a person IRI               
for idx, value in df.iterrows():
    df['IRI'][idx] = 'person_' + str(idx)

#Add IRIs to the normalized and unnormalized occupations
for i in range(0, len(df.IRI)):
    if df['Occupation'][i] !='':
        if df['Occupation'][i] not in occupations_dict.keys():
            df['Occ_IRI'][i] = 'occupation_' + str(occ_idx)
            occ_idx += 1
            occupations_dict.update({df['Occupation'][i]: df['Occ_IRI'][i]})
        else:
            df['Occ_IRI'][i] = occupations_dict.get(df['Occupation'][i])

    if df['Occupation 1'][i] !='':
        if df['Occupation 1'][i] not in occupations_dict.keys():
            df['Occ_1_IRI'][i] = 'occupation_' + str(occ_idx)
            occ_idx += 1
            occupations_dict.update({df['Occupation 1'][i]: df['Occ_1_IRI'][i]})
        else:
            df['Occ_1_IRI'][i] = occupations_dict.get(df['Occupation 1'][i])

    if df['Occupation 2'][i] !='':
        if df['Occupation 2'][i] not in occupations_dict.keys():
            df['Occ_2_IRI'][i] = 'occupation_' + str(occ_idx)
            occ_idx += 1
            occupations_dict.update({df['Occupation 2'][i]: df['Occ_2_IRI'][i]})
        else:
            df['Occ_2_IRI'][i] = occupations_dict.get(df['Occupation 2'][i])

    if df['Normalized Occupation'][i] !='':
        if df['Normalized Occupation'][i] not in occupations_dict.keys():
            df['Norm_Occ_IRI'][i] = 'occupation_' + str(occ_idx)
            occ_idx += 1
            occupations_dict.update({df['Normalized Occupation'][i]: df['Norm_Occ_IRI'][i]})
        else:
            df['Norm_Occ_IRI'][i] = occupations_dict.get(df['Normalized Occupation'][i])

    if df['Normalized Occupation 1'][i] !='':
        if df['Normalized Occupation 1'][i] not in occupations_dict.keys():
            df['Norm_Occ_1_IRI'][i] = 'occupation_' + str(occ_idx)
            occ_idx += 1
            occupations_dict.update({df['Normalized Occupation 1'][i]: df['Norm_Occ_1_IRI'][i]})
        else:
            df['Norm_Occ_1_IRI'][i] = occupations_dict.get(df['Normalized Occupation 1'][i])

    if df['Normalized Occupation 2'][i] !='':
        if df['Normalized Occupation 2'][i] not in occupations_dict.keys():
            df['Norm_Occ_2_IRI'][i] = 'occupation_' + str(occ_idx)
            occ_idx += 1
            occupations_dict.update({df['Normalized Occupation 2'][i]: df['Norm_Occ_2_IRI'][i]})
        else:
            df['Norm_Occ_2_IRI'][i] = occupations_dict.get(df['Normalized Occupation 2'][i])
       
df.head(195)
#occupations_dict
```
