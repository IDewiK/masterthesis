## Creating the DataFrame with information about companies

The DataFrame is created in several steps. Details about each steps are explained in the corresponding code cells. Similar to person entries separate lists with all the necessary information about compnaies are extracted. When all the necessary information are collected these lists are merged into a DataFrame. The DataFrame is later saved and extracted as a csv table.

```python
'''Identifying the HouseOwnership and adding True if a person 
is a houseowner and False if the person is not a house owner'''
HouseOwnerSymbol = ['*']
new = []

for i in range(0, len(company_entry_list)):
    if '*' in company_entry_list[i]:
        company_entry_list[i] = company_entry_list[i][:] + ', True'
        new.append(company_entry_list[i])
    else:
        company_entry_list[i] = company_entry_list[i][:] + ', False'
        new.append(company_entry_list[i])

company_entry_list = new
#person_entry_list
company_entry = '\r\n'.join(company_entry_list)
company_entry_list
```

```python
'''Deleting the * symbol after adding "True/False" at the end of each entry'''
company_entry = '\r\n'.join(company_entry_list)
company_entry = re.sub(r'\*\s', '', company_entry)
company_entry = re.sub('(\d+\,)(\s[A-ZÄÖÜa-zäöüß]+)\.', r'\1\2,', company_entry)
company_entry = re.sub('(\d+\,)(\s[A-ZÄÖÜa-zäöüß]+)(\s)(\w)', r'\1\2,\3\4', company_entry)
company_entry_list = company_entry.split('\r\n')
company_entry_list
```

```python
'''Creating list of street names as a basis for the dataframe'''
company_entry_list = company_entry.split('\r\n')
testlist = company_entry_list[:30]
streetlist = []
company_entry_wo_streets = []

for i in company_entry_list:
    iSplit = i.split(', ')
    streetlist.append(iSplit[0])
    rest = ', '.join(iSplit[1:])
    company_entry_wo_streets.append(rest)
    #print(rest)
#streetlist
```

```python
'''Creating list of house numbers and adding the house numbers into the company_df'''
#pd.set_option('max_columns', None)
company_df = pd.DataFrame()
company_df['House Number'] = ''

housenumberList = []
company_entry_wo_housenumber = []
companynameList = []
company_entry_wo_companyname = []

for i in company_entry_wo_streets:
    iSplit = i.split(', ')
    housenumberList.append(iSplit[0])
    rest = ', '.join(iSplit[1:])
    company_entry_wo_housenumber.append(rest)
#company_df = pd.DataFrame({'Street': streetlist,'House Number': housenumberList})


houseownerList = []
company_entry_wo_name_owner = []
for i in company_entry_wo_housenumber:
    iSplit = i.split(', ')
    houseownerList.append(iSplit[-1])
    rest = ' '.join(iSplit[:-1])
    company_entry_wo_name_owner.append(rest)
    
#print(company_entry_wo_name_owner)

#company_df = pd.DataFrame({'Street': streetlist, 'House Number': housenumberList, 'House Owner': houseownerList, 'Company Name': company_entry_wo_name_owner})    

OwnercompanyLocationList = []
LocationStreetList = []
LocationNumberList = []
CompanyNameList = []


for i in company_entry_wo_name_owner:
    location_match = re.search(r'(([A-Za-zäöüß\-]*\s*(st[rt](\.|aß[ce])|platz|gasse|weg|br[uü]cke|halle|markt)|[A-Za-zäöüß\-]*\s*(Straße|Platz|Gasse|Weg|Br[uü]cke|Halle|Bahnlinie|Markt))\s[0-9/]+)', i)
    if location_match:
        OwnercompanyLocationList.append(location_match[0])
    else:
        OwnercompanyLocationList.append('')
        
for i in OwnercompanyLocationList:
    iSplit = i.split(' ')
#    if iSplit[-1].isdigit():
    if any(j.isdigit() for j in iSplit[-1]):
        LocationNumberList.append(iSplit[-1])
        rest = ' '.join(iSplit[:-1])
    else:
        LocationNumberList.append('')
        rest = ' '.join(iSplit[0:])
    LocationStreetList.append(rest)

for i in company_entry_wo_name_owner:
    companyNames = re.sub(r'(([A-Za-zäöüß\-]*\s*(st[rt](\.|aß[ce])|platz|gasse|weg|br[uü]cke|halle|markt)|[A-Za-zäöüß\-]*\s*(Straße|Platz|Gasse|Weg|Br[uü]cke|Halle|Bahnlinie|Markt))\s[0-9/]+\.*)', '', i)
    if len(companyNames) > 0:
        CompanyNameList.append(companyNames)
    else:
        CompanyNameList.append('')
        
company_df = pd.DataFrame({'Street': streetlist, 'House Number': housenumberList, 'House Owner': houseownerList, 'Temp': CompanyNameList, 'Owner Street': LocationStreetList, 'Owner Number': LocationNumberList}) 
company_df.head(45)
#company_entry_wo_name_owner
```

```python
'''Extraction of additional information and some corrections within the Company Name column'''
#extract additional information to another column and delete it in the company name
company_df['Temp'] = company_df['Temp'].str.replace(r'\((\s?C\.?|Cpt\.?|Cptr\.?|Compt\.?|Campt\.?)\)','(Comptoir)')
company_df['Temp'] = company_df['Temp'].str.replace(r'\((L\.?|Ld\.?|Lad\.?)\)','(Laden)')
company_df['Temp'] = company_df['Temp'].str.replace(r'\(Fil\.?\)','(Filiale)')
company_df['Temp2'] = company_df['Temp'].str.extract(r'\b(HIV\b|HI\b|HII\b|HIII\b|H0\b|HO\b|O\b|0\b|IV\b|V\b|I\b|II\b|III\b)')
company_df['Temp2'] = company_df['Temp2'].fillna('')
company_df.fillna('')
#separate House Floor from Building Part
company_df['Part of House'] = np.where(company_df['Temp2'].str.contains(r'H'), 'Hinterhaus', 'Vorderhaus')
company_df['House Floor'] = np.where(company_df['Temp2'].str.contains(r'\b0\b|H0\b|\bO\b|HO\b'), '0', 
                            np.where(company_df['Temp2'].str.contains(r'\bI\b|HI\b'), '1', 
                            np.where(company_df['Temp2'].str.contains(r'\bII\b|HII\b'), '2', 
                            np.where(company_df['Temp2'].str.contains(r'\bIII\b|HIII\b'), '3', 
                            np.where(company_df['Temp2'].str.contains(r'\bIV\b|HIV\b'), '4', 
                            np.where(company_df['Temp2'].str.contains(r'\bV\b|HV\b'), '5',''))))))

#Create column with Full Address and Full Owner Address
company_df['Full Address'] = company_df['Street'] + ' ' + company_df['House Number']
company_df['Building'] = company_df['Street'] + ' ' + company_df['House Number']
company_df['BuildingPart'] = company_df['Street'] + ' ' + company_df['House Number'] + ', ' + company_df['Part of House']
company_df['BuildingPartFloor'] = company_df['Street'] + ' ' + company_df['House Number'] + ', ' + company_df['Part of House'] + ', ' + company_df['House Floor']
company_df['BuildingPartFloor'] = company_df['BuildingPartFloor'].str.rstrip(' , ')
company_df['Full Owner Address'] = company_df['Owner Street'] + ' ' + company_df['Owner Number']
company_df['Owner Building'] = company_df['Owner Street'] + ' ' + company_df['Owner Number']

#delete the location address in the company name that has already been transferred to the Location Owner Street and Number columns
company_df['Temp'] = company_df['Temp'].str.replace(r'(([A-Za-zäöüß\-]*\s*(st[rt](\.|aß[ce])|platz|gasse|weg|br[uü]cke|halle|markt|thal)|[A-Za-zäöüß\-]*\s*(Straße|Platz|Gasse|Weg|Br[uü]cke|Halle|Bahnlinie|Markt))\s[0-9/]+)','')
company_df['Temp'] = company_df['Temp'].str.replace(r'\b(HIV|H(I){1,3}|H0|HO|O|0|U|M|W|IH|V|N|IV|(I){1,3}|\d+[a-g])\b','')
company_df['Temp'] = company_df['Temp'].str.replace(r'^[^A-Z]+','')
company_df['Temp'] = company_df['Temp'].str.replace(r'\s\.','.')
company_df['Temp'] = company_df['Temp'].str.replace(r'\.\s\.\s','. ')
company_df['Temp'] = company_df['Temp'].str.replace(r'\s[A-ZÄÖÜa-uw-zäöüß]\s',' ')

#extracting the industry of a company
company_df['Industry'] = company_df['Temp'].str.extract(r'(Agentur|Aktiengesellschaft|Brauerei\sA\.\-G\.|A\.\-G\.?|[A-Za-zäöüß]+anstalt|Anstalt|Apotheke|[A-Za-zäöüß]+apotheke|Atelier|[A-Za-zäöüß]*[Bb]{1}ank|Bankgeschäft|Akt[\.|\,]\-Ges\.|Brauhaus|[A-ZÄÖÜa-zäöüß]*[Bb]{1}rauerei\b|Bureau\b|Comptoir|Fabrik\s|[A-Za-zäöüß]*[Ff]{1}abrik\b|Fabrikbureau\b|Filiale|[A-ZÄÖÜa-zäöüß]*\.?\s?[Ff]{1}irma|G\.\sm\.\sb\.\sH\.|[A-Za-zäöüß]*\.\sGeschäft\b|[A-Za-zäöüß]+\.\-Geschäft\b|[A-Za-zäöüß]+\.\-Gesch\.|[A-Za-zäöüß]*[Gg]{1}eschäft|[A-Za-zäöüß]*[Gg]{1}esellschaft\b|[A-Za-zäöüß]+\-\sund\s[A-Za-zäöüß]+handlung|[A-Za-zäöüß]+\-\s*u\.\s*[A-Za-zäöüß]+[Hh]{1}andlung|[A-Za-zäöüß]+handlung|Handlung|[A-Za-zäöüß]*hdlg\.?|[A-Za-zäöüß]+\-\su\.\s[A-Za-zäöüß]*handlg\.?|[A-Za-zäöüß]*\-\su\.\s[A-Za-zäöüß]*hdlg\b\.?|[A-Za-zäöüß]*\-\su\.\s[A-Za-zäöüß]*handlg\b\.?|G\.\-L\.?|G\.\sL\.|GL\b|Kaufl\.|Laden\b|[A-Za-zäöüß]*[Ll]{1}ager\b|Ges\.\-Firma|[A-Za-zäöüß]*\.?\-?Malerei|[CZ]{1}i[gq]{1}arrenh\.|[A-Za-zäöüß]+zollamt)')
company_df['Company Name'] = company_df['Temp'].str.replace(r'(Agentur|Aktiengesellschaft|A\.\-G\.?|\(?[A-Za-zäöüß]*anstalt\)?|Anstalt|Apotheke|\(?[A-Za-zäöüß]+apotheke\)?|\(?Atelier\)?|Bankgeschäft|Akt[\.|\,]\-Ges\.|[A-ZÄÖÜa-zäöüß]*\(?[Bb]{1}rauerei\b\)?|\(Bureau\)|\(?Fabrikbureau\b\)?|\(C.\)|\(Comptoir\)|\(?Fabrik\)?\s|[A-Za-zäöüß\-]\s*und\s*[A-Za-zäöüß]*[Ff]{1}abrik\b|\(?[A-Za-zäöüß]*[Ff]{1}abrik\b\)?|\(Fabrikbureau\)|\(?Filiale\)?|[A-ZÄÖÜa-zäöüß]*\.?\s?[Ff]{1}irma|\(?G\.\sm\.\sb\.\sH\.\)?|[A-Za-zäöüß]*\.\sGeschäft\b|[A-Za-zäöüß]*[Gg]{1}eschäft|[A-Za-zäöüß]+\.\-Geschäft\b|[A-Za-zäöüß]+\.\-Gesch\.|[A-Za-zäöüß]*[Gg]{1}esellschaft\b|[A-Za-zäöüß]+\-\sund\s[A-Za-zäöüß]+[Hh]{1}andlung|[A-Za-zäöüß]+\-\s*u\.\s*[A-Za-zäöüß]+[Hh]{1}andlung|Handlung|[A-Za-zäöüß]*hdlg\.?|[A-Za-zäöüß]+\-\su\.\s[A-Za-zäöüß]+handlg\.?|[A-Za-zäöüß]+handlg\.?|[A-Za-zäöüß]*\-\su\.\s[A-Za-zäöüß]+hdlg\b\.?|[A-Za-zäöüß]*\-\su\.\s[A-Za-zäöüß]*handlg\b\.?|\(?[A-Za-zäöüß]+lager\)?|\(?Lager\b\)?|\(?G\.\-L\.?\)?|\(?G\.\sL\.\)?|\(?GL\b\)?|\(?L\.\)?|\(Lad\.\)|\(?Laden\b\)?|\(?Kaufl\.\)?|Ges\.\-Firma|[A-Za-zäöüß]*\.?\-?Malerei|\(?[CZ]{1}i[gq]{1}arrenh\.\)?|\(?[A-Za-zäöüß]+zollamt\)?)\s*', '')

company_df.applymap(lambda x: x.rstrip() if type(x)==str else x)

company_column_order = ['Company Name', 'Industry', 'Full Address', 'Street', 'House Number', 'Temp2', 'Part of House', 'House Floor', 'Building', 'BuildingPart', 'BuildingPartFloor', 'House Owner', 'Full Owner Address', 'Owner Street', 'Owner Number', 'Temp']
company_df = company_df.reindex(columns=company_column_order)
company_df.head(50)
```
